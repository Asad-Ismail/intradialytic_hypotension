```python
all_cols=merged_df.columns.tolist()
```

```python
hypotension_cols=[item for item in all_cols if 'hypotension' in item]
```

```python
hypotension_cols
```

```python
df_rm_idhs=merged_df.drop(columns=hypotension_cols+['patient_id'],axis=1)
```

```python
df_rm_idhs.columns
```

```python
select_cols=df_rm_idhs.select_dtypes(include=['int64', 'float64', 'bool']).columns.tolist()
```

```python
df_cols_idhs=merged_df[hypotension_cols]
```

```python
df_cols_idhs.columns
```

```python
## Define columns for correlation
#df1_cols = ['dialysis_lengh_in_m','has_diabetes']
df1_cols=select_cols
df2_cols = ['is_hypotension_drop_ge_20','is_hypotension_nadir_lt_90'] 
```

```python
correlation_matrix = pd.DataFrame(index=df1_cols, columns=df2_cols)

# Calculate the correlations
for col1 in df1_cols:
    for col2 in df2_cols:
        correlation = df_rm_idhs[col1].corr(df_cols_idhs[col2])
        correlation_matrix.loc[col1, col2] = correlation

# Convert the correlation matrix to numeric type for plotting
correlation_matrix = correlation_matrix.astype(float)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation between Columns')
plt.ylabel('df1 Columns')
plt.xlabel('df2 Columns')
plt.show()
```

```python
correlation_matrix
```

```python
max_corr = correlation_matrix.max()
```

```python
n=5
corr_series = correlation_matrix.unstack().dropna().sort_values()

# Get top n and bottom n correlations
top_n = corr_series.tail(n)
bottom_n = corr_series.head(n)

# Print top n correlations
print(f"Top {n} correlations:")
for (idx1, idx2), val in top_n.items():
    print(f"{idx1} and {idx2}: {val:.4f}")
```

```python
print(f"Bottom {n} correlations:")
for (idx1, idx2), val in bottom_n.items():
    print(f"{idx1} and {idx2}: {val:.4f}")
```
